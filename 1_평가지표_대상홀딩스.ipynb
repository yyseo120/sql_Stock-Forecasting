```python
# 전체 종목의 총주식수를  수집하기 전에 일부 종목을 먼저 수집합니다.
item_code = "084690"
item_name = "대상홀딩스"
```

```python
# 종목상세정보를 위한 url을 확보합니다.

item_url = f"https://finance.naver.com/item/main.nhn?code={item_code}"
item_url
```

```python
# read_html을 이용해 데이터를 수집합니다.
tables = pd.read_html(item_url, encoding="cp949")
len(tables)
```

```python
# 수집한 테이블들 출력
for table in tables:
    print("="*100)
    print(table)
    print("="*100)
```

```python
# tables [5]에 상장 주식수가 저장되 있음
tables[5]
```

```python
# tables[5] 의 2번째줄 1번째 칸의 데이터 (상장주식수)
tables[5].iloc[2 ,1]
```

```python
# 상장 주식수를 리턴하는 함수.
import time

def get_all_stock(item_code):
    # 종목상세정보를 위한 url을 확보합니다.  
    item_url = f"https://finance.naver.com/item/main.nhn?code={item_code}"
    # read_html을 이용해 데이터를 수집합니다.
    tables = pd.read_html(item_url, encoding="cp949")
    # tables[5] 의 2번째줄 1번째 칸의 데이터 (상장주식수)
    all_stock = tables[5].iloc[2 ,1]
    # 0.1초 멈춤   
    time.sleep(0.1)
    # 상장 주식수를 정수로 변환해서 리턴
    return int(all_stock)
```

```python
# 상장 주식수 조회
get_all_stock("000660")
```

```python
# tqdm은 total quality data management의 약자입니다.
# tqdm.auto 에서 tqdm을 불러옵니다.
from tqdm.auto import tqdm
#DataFrame에 tqdm 함수 사용가능하도록 설정
tqdm.pandas()
```

```python
# DataFrame의 종목코드 column에 있는 종목코드들값을 get_all_stock 함수에 적용되어 상장주식수를 조회
# result에 series형태로 저장됩니다.
result = df["STK_CD"].progress_apply(get_all_stock)
```

```python
result
```

```python
# result에 저장된 상장 주식수를 컬럼 상장주식수에 대입
df["상장주식수"] = result
```

```python
df
```

```python
# 반도체 종목의 당기 순이익 조회
sql = """
      SELECT T1.STK_CD, T1.STK_NM, T1.SEC_NM, T1.EX_CD
              ,T3.FIN_ITM_NM, T3.FIN_ITM_VAL, T3.DAT_SOU
      FROM stock T1
      INNER JOIN FINANCE_Y T3
      ON (T1.STK_CD = T3.STK_CD)
      WHERE T1.SEC_NM = '음식료'
      AND T1.EX_CD = 'KP'
      AND T3.FIN_ITM_NM = '당기순이익'
      AND T3.YY = '2019';
"""  
```

```python
result
```

```python
df2
```

```python
# FIN_ITM_VAL 컬럼의 타입을 double로 변환
df2["FIN_ITM_VAL"] = df2["FIN_ITM_VAL"].astype('double')
```

```python
# df 데이터 프레임과 df2의 "STK_CD","FIN_ITM_VAL" 병합
df = pd.merge(df, df2[["STK_CD","FIN_ITM_VAL" ]])
```

### 3. BPS

```python
# 전체 종목의 총주식수를  수집하기 전에 일부 종목을 먼저 수집합니다.
item_code = "000660"
item_name = "SK하이닉스"
```

```python
# 종목상세정보를 위한 url을 확보합니다.

item_url = f"https://finance.naver.com/item/main.nhn?code={item_code}"
item_url
```

```python
# read_html을 이용해 데이터를 수집합니다.
tables = pd.read_html(item_url, encoding="cp949")
len(tables)
```

```python
# 수집한 테이블들 출력
for table in tables:
    print("="*100)
    print(table)
    print("="*100)
```

```python
# tables [3]에 BPS가 저장되 있음
tables[3]
```

```python
# tables[3] 의 11번째줄 1번째 칸의 데이터 (2018년 BPS)
tables[3].iloc[11 ,1]
```

```python
# BPS 를 리턴하는 함수.
import time

def get_bps(item_code):
    # 종목상세정보를 위한 url을 확보합니다.  
    item_url = f"https://finance.naver.com/item/main.nhn?code={item_code}"
    # read_html을 이용해 데이터를 수집합니다.
    tables = pd.read_html(item_url, encoding="cp949")
    # tables[3] 의 11번째줄 1번째 칸의 데이터 (2018년 BPS)
    bps = tables[3].iloc[11 ,1]
    # 0.1초 멈춤   
    time.sleep(0.1)
    # BPS 를 정수로 변환해서 리턴
    return int(bps)
```

```python
# bps 조회
get_bps("000660")
```

```python
# tqdm은 total quality data management의 약자입니다.
# tqdm.auto 에서 tqdm을 불러옵니다.
from tqdm.auto import tqdm
#DataFrame에 tqdm 함수 사용가능하도록 설정
tqdm.pandas()
```

```python
# DataFrame의 종목코드 column에 있는 종목코드들값을 get_bps 함수에 적용되어 bps를 조회
# result에 series형태로 저장됩니다.
result = df["STK_CD"].progress_apply(get_bps)
```

```python
result
```

```python
# BPS 컬럼에 대입
df["BPS"] = result
```

```python
df
```

```python
# 종목코드, 종목명, BPS 조화
df[["STK_CD", "STK_NM", "BPS"]]
```
